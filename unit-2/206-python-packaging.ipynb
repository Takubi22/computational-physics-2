Packaging python code involves creating an installable python package.

The python packaging system is constantly evolving, and the current recommendations of tools is list here:

https://packaging.python.org/en/latest/guides/tool-recommendations/

##  Directory structure: 

We will work on the example provided in the above course that builds on the Mandelbrot set exercise.  

On your local computer, if you have `git` installed, you can clone this via:

```
git clone https://github.com/sbu-python-class/mymodule.git
```

The directory structure appears as:

```
mymodule
├── mymodule
│   ├── __init__.py
│   └── mandel.py
└── README.md
└── pyproject.toml (previously setup.py)
```

### Package structure:

* The top-level `mymodule` directory is not part of the python package, but instead is where the source control (e.g. git) begins, and also hosts setup files that are used for installation.
  

* `mymodule/mymodule` is the actual python module that we will load. To make python recognize this as a module, we need an `__init__.py` file there -- it can be completely empty.
   

* The actual `*.py` files that make up our module are in `mymodule/mymodule`


### Command line access:
This package does not appear in our python search path, so the only way to load it is to work in the top-level `mymodule/` directory:

```
cd mymodule
```

And then we can do:

```python
import mymodule.mandel
fig = mymodule.mandel.mandelbrot(256)
fig.savefig("test.png")
```

we could also do:

```python
from mymodule.mandel import mandelbrot
```

## Using `setuptools`

`setuptools` is used to package our library.  See the packaging guidelines here:

https://packaging.python.org/en/latest/guides/distributing-packages-using-setuptools/

The main thing we need to do is create a `setup.py` that describes our
package and its requirements:

Here's a first `setup.py`:

#### You may edit it with VIM: 

`vim pyproject.toml`

```python
[build-system]
requires = ["setuptools"]
build-backend = "setuptools.build_meta"

[project]
name = "mymodule"
description = "test module"
readme = "README.md"
license.text = "BSD"
version="0.1.0"
authors = [
  {name="Your Name"},
  {email="Your Email"},
]

dependencies = [
  "numpy",
  "matplotlib",
]
```

**Note:**

See the description of the BSD licence here:

https://en.wikipedia.org/wiki/Berkeley_Software_Distribution


# How to setup code?

We can setup in a variety of ways.  Two useful ways are:

## Install and uninstall with pip:

Inside the module directory:

  `python -m pip install .`


## Development mode:

Install in Editable Mode (for development):

  `python -m pip install -e .`
  

  `pip install -e .`


With modern pip + setuptools, we follow PEP 660 editable installs. This does NOT necessarily create OS-level symlinks. Instead, it installs a special loader that points Python to your source directory. All changes to .py files are reflected immediately. It behaves like a symlink from Python’s perspective, but may not literally create one at the filesystem level.
  
## After installation:

This will copy the source files into your conda environment putting them into your python search path. Then you can use this package from anywhere. For instance, in my computer mymodule is installed here:

    >>> print(site.getsitepackages())
