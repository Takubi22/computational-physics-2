## Scaling plot:

````python
# Fix parameter
max_iter = 20

def run(N):
    """
    Function to compute execution times for different problem sizes (N).
    """
   # Add a time stamp 
    start = time.time()
    #print(start)

    # Call the md function
    m = md.mandelbrot(N, max_iter = max_iter)
    
    # Add another time stamp
    end   = time.time()
    #print(end)

    # Save the execution time:
    exec_time = end - start
    #print(f"The execution time was = {end - start} s.\n")

    return exec_time
````

````python
import numpy as np
````

````python
# List to loop over
n_list = [64, 128, 256, 512, 1024, 2048]

# Empty list to store the execution time
e_list = []

# For loop to call our function
for j in n_list:
    #print(j)
    e_j = run(j)
    e_list.append(e_j)

#print(e_list)   

# Convert to arrays:
n_list = np.array(n_list)
e_list = np.array(e_list)
````

# Scaling Figure

````python
import matplotlib.pyplot as plt
````

````python
# Figure environment

plt.figure(figsize=(11,3))

plt.plot(n_list, e_list)

plt.xlabel(r"$N$")
plt.ylabel(r"$t_N$")

plt.show()
````

````python
# Figure environment

plt.figure(figsize=(11,3))

plt.plot(np.log2(n_list), np.log2(e_list))

plt.xlabel(r"$\log_2(N)$")
plt.ylabel(r"$\log_2(t_N)$")

plt.show()
````
